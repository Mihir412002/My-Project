```python
X = iris.drop(['species'],axis=1)
```

```python
Y=iris['species']
```

```python
le = LabelEncoder()
Y = le.fit_transform(Y)
Y
```

```python
#we get its encoding as above sentosa:0,versicolor:1,verginica:2
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

Logistic Regression

```python
lr = LogisticRegression(solver='newton-cg')
lr.fit(X_train,Y_train)
```

Random forest is ensemble or collection of desicion tree

```python
Y_pred1 = lr.predict(X_test)
```

```python
#get the confusion matrix
confusion_matrix(Y_test,Y_pred1)
```

```python
sns.heatmap(data=confusion_matrix(Y_test,Y_pred1), annot=True, cmap='Blues')
plt.ylabel = ('Actual Label')
plt.xlabel('Predict label')
```

Get the accuracy score

```python
accuracy_score(Y_test,Y_pred1)
```

Decision Tree

```python
dtree = DecisionTreeClassifier()
dtree.fit( X_train,Y_train)
```

```python
Y_pred2 = dtree.predict(X_test)
```

Get the confusion matrix,heatmap for confusion matrix & accuracy score

```python
confusion_matrix(Y_test,Y_pred2)
```

```python
sns.heatmap(data=confusion_matrix(Y_test,Y_pred2), annot=True, cmap='Blues')
plt.ylabel = ('Actual Label')
plt.xlabel('Predict label')
```

```python
accuracy_score(Y_test,Y_pred2)
```

```python
plt.figure(figsize = (20,20))
dec_tree = plot_tree(decision_tree = dtree,feature_names = iris.columns[:-1],class_names = ["setosa","vercicolor","verginica"],filled=True)
```
